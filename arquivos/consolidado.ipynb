```python
import pandas as pd
import pdfplumber
import openpyxl
```

```python
# Caminho para o arquivo PDF e para o arquivo Excel de saída usando string bruta
pdf_path = r'C:\github\modulo_python_digital\arquivos\fund.pdf'
output_xlsx_path = r'C:\github\modulo_python_digital\arquivos\dados_extraidos.xlsx'

# Lista para armazenar dados extraídos
data = []

# Abre o PDF com pdfplumber
with pdfplumber.open(pdf_path) as pdf:
    # Itera sobre cada página do PDF
    for page in pdf.pages:
        # Extrai as tabelas da página
        tables = page.extract_tables()
        for table in tables:
            # Adiciona as tabelas extraídas à lista de dados
            data.extend(table)

# Converte a lista de dados em um DataFrame do pandas
df = pd.DataFrame(data)

# Salva DataFrame em um arquivo Excel, mantendo a formatação básica
with pd.ExcelWriter(output_xlsx_path, engine='xlsxwriter') as writer:
    df.to_excel(writer, sheet_name='Sheet1', index=False)

print(f"Arquivo Excel salvo como {output_xlsx_path}")
```
